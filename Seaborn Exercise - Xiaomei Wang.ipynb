```python
# exercise 1
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
# exercise 2
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species', height=3)
```

```python
# exercise 3
iris = sns.load_dataset('iris')
plt.figure(figsize=(10, 8))
sns.boxplot(data=iris)
```

```python
# exercise 4
species_colors = ['#78C850', '#F08030', '#6890F0', '#a020f0' ]
sns.violinplot( data=iris, palette=species_colors)
```

```python
# exercise 5
sns.histplot(iris, x='petal_length', kde=True)
```

```python
penguins = sns.load_dataset("penguins")
sns.histplot(data=penguins, x="flipper_length_mm")
```

```python
# exercise 6
sns.swarmplot(y='petal_length', x='species', data=iris, palette=species_colors, hue='species', legend=False)
```

```python
# Exercise 7
sns.barplot(penguins, x="island", y="body_mass_g")
```

```python
# exercise 8
df = pd.DataFrame(iris)
correlation = df.corr(numeric_only=True)
sns.heatmap(correlation, annot = True, square = True)
```
